## Definitions

Simplified definition for components:

   * Input: $N$ data points, $X=[x_1,x_2,\ldots, x_N]$, where $x_i \in D^n$ ($D$ is the domain depending on your application and similarity function).
   This is an abstract notation and is already pre-processed, e.g. shingling for texts, color histogram for images, ...
   * Similarity: feature domain $D^n$, a similarity measure $s(x,y)$ for $x,y \in D^n$.
   * Hash family $H$: for random $h \in H$, $Pr[h(x)=h(y)] \approx s(x,y) $.
   
We discussed three similarity measures:

   * $D=\{0,1\}$ and $s$ is jaccard coefficient: use MinHash as $H$.
   * $D=\mathbb{R}$ and $s(x,y)=cos(x,y)$: use random hyperplane as $H$.
   * $D=\mathbb{R}$ and $d(x,y)$ is the Euclidean distance (negatively related with $s$): use random lines as $H$.
   

## Bootstrap single hash (linear probability to S-curve)

Say we already get a random hash $Pr[h(x)=h(y)]=s(x,y)$.

Now we want to find highly similar items via (a set of) $h$.
Let $t \in [0,1]$ be the threshold for our target similarity.

   * Ground-truth similarity:
   If $s(x,y)\ge t$, output YES; else output NO.
   * Predicted similarity: 
   If $h(x)=h(y)$, output YES; else output NO.

Let $p(t)$ be the shorthand of $Pr[h(x)=h(y)|s(x,y) \ge t]$.
From the setup, we know $p(t) = Pr[h(x)=h(y)|s(x,y) \ge t] = t$.
Our first application involves the evaluation of only one $h$.
We plot so the overall probability to output YES $P(t)=p(t)$ with different $t$:

```python
t = arange(0,1,0.01)
p = t
P = p
plot(t, P)
```

This has a problem: it tends to give false positives. That is, even not so similar pairs ($s(x,y)=0.5$) have non-trivial probability to be predicted "YES". Our objective is to bend the curve downward.

The "**AND**" trick,i.e., multiple rows one band in the lecture:

   * Suppose we apply $r$ such random $h$. 
   * We output YES iff. all $r$ such random $h$ predicts "YES". (i.e. "AND").

If the pair has low similarity, e.g. $s(x,y)=0.5$, it has much less chance, e.g. $0.5^r$,to be predicted YES by our overall construction. More precisely:

$$P(t)=p(t)^r$$

We plot it:

```python
t = arange(0,1,0.01)
def prob_and(r, t):
    p = t
    P = p ** r
    return P
r = range(1, 5)
PP = [prob_and(x, t) for x in r]
tt = [t] * 4
# the convention of plot is use each column vector,
# so we transpose them first
plot(matrix(tt).T, matrix(PP).T)
axis([0, 1, 0, 1])
legend(['r=%d' % x for x in r], loc='upper left')
```

Now the problem of false **positive** is well solved.
For example, when $r=4$, a pair with $s(x,y)=0.5$ will only be predicted YES with a probability smaller than $0.1$.

The new problem is false **negative**.
Observe the pairs with $s(x,y)=0.8$, the overall probability to predict YES becomes lower than $0.4$ after the AND operation.
We can use "OR" trick to alleviate the problem:

   * We call the previous AND construction as an $r$-group.
   * We apply $B$ such $r$-group, i.e.,multiple bands.
   * We predict YES iff. at least 1 out of the $B$ $r$-groups predicts YES.
   
Formally, the overall probability is:

$$P(t)=1-(1-p(t)^r)^B$$

Let's visualize this:

```python
t = arange(0,1,0.01)
def prob_and_or(r, B, t):
    p = t
    P = 1 - (1 - p ** r) ** B
    return P

def prob_visualize(range_r, range_B):
    import itertools
    # Generate all the combinations of parameters
    params = list(itertools.product(range_r, range_B))
    PP = [prob_and_or(x, y, t) for (x, y) in params]
    tt = [t] * (len(range_r) * len(range_B))
    # the convention of plot is use each column vector,
    # so we transpose them first
    plot(matrix(tt).T, matrix(PP).T)
    axis([0, 1, 0, 1])
    legend(['r=%d, B=%d' % (x, y) for (x, y) in params], loc='upper left')
    
prob_visualize(range(1,2), range(1,5))
```

We already see that by applying AND, OR separately, we can bend the curve upwards and downwards.
You may try other parameters of $r$ and $B$ ("rows" and "bands" in the lecture slides) and target any similarity threshold.
For example:

```python
prob_visualize(range(8,10), range(8,10))
```

## Amplify probability gap for any function

Note, for a single $h$ to be a useful building block.
The probability $Pr[h(x)=h(y)]$ should be positively related with similarity $s(x,y)$.
In the last section of lecture slides, we discussed based on distance.
Distance is a kind of dissimilarity, which is negatively related with $s(x,y)$.
We only discuss similarity here, but the principle is the same for distance.

The requirement of "positively related" here is less stringent than "monotonic increasing".
It works as long as there is a gap.
Say our single $h$ can yield a probability like this: 

```python
plot([0,0.4,0.6,1], [0,0.3,0.7,1])
```

It's just an illustration.
The linear response of MinHash is a special instance of this type of function.
The essense is that you can find two control points:

   * $(s_1, p_1)$: for all $(x, y)$ that $s(x,y) \le s_1$, we have $Pr[h(x)=h(y)] \le p_1$
   * $(s_2, p_2)$: for all $(x, y)$ that $s(x,y) \ge s_2$, we have $Pr[h(x)=h(y)] \ge p_2$

In the above illustration, $(s_1, p_1)=(0,4, 0.3)$ and $(s_2, p_2)=(0.6, 0.7)$.

We use the same AND-OR trick.

We first argue the YES output.
With parameters $r$ and $B$.
It is same as previous analysis:

$$Pr[h(x)=h(y) | s(x,y) \ge s_2] = 1-(1-p(t)^r)^B \ge 1-(1-p_2^r)^B $$

Then we argue the NO output.

$$Pr[h(x)=h(y) | s(x,y) \le s_1] = 1-(1-p(t)^r)^B \le 1-(1-p_1^r)^B $$

Note that one is giving upper bound and the other is given lower bound.
Let's visualize the tow bounds.

```python
def prob_and_or_p(r, B, p):
    return 1 - (1 - array([p] * len(t)) ** r) ** B

def visualize_amplification(r, B, s_1, p_1, s_2, p_2):
    upper_s_1 = prob_and_or_p(r, B, p_1)
    lower_s_2 = prob_and_or_p(r, B, p_2)
    figure()
    plot([0, s_1], [upper_s_1, upper_s_1], 'b')
    hold(True)
    plot([s_2, 1], [lower_s_2, lower_s_2], 'b')
    # enlarge the axis a bit so that the curves are not masked by bounding box
    axis([-0.1, 1.1, -0.1, 1.1]) 

# The gap before amplification
visualize_amplification(1, 1, 0.4, 0.3, 0.6, 0.7)
# Let's try something.
visualize_amplification(10, 10, 0.4, 0.3, 0.6, 0.7)
# Oops, the above does not give YES instance enough probability
# Try other r and B
visualize_amplification(7, 50, 0.4, 0.3, 0.6, 0.7)
```

Note the last plot, it (visually) give an ideal amplification.
But how to find such good parameter of r and B?

It can be done by solving the equation array:

$$1-(1-p_2^r)^B \ge 0.99 $$
$$1-(1-p_1^r)^B \le 0.01 $$

**NOTE**:

   * After amplification, the region between our two control points is not defined.
   We don't know the exact probability.
   We can only say something for the two ends.
   * For the two ends, we also don't know the **exact** probability.
   What we know is the upperbound and lowerbound.
   This is good enough for our application: determine whether two items are similar or not.

## A synthesized data set

```python
import numpy as np
def gen_one_cluster(center, stdev, n):
    x_0, y_0 = center
    x = x_0 + stdev * randn(1, n)
    y = y_0 + stdev * randn(1, n)
    m = matrix(concatenate((x,y), axis=0))
    # each column is a data point
    return m

X_1 = gen_one_cluster((-1, 0), 0.2, 10)
X_2 = gen_one_cluster((1, 0), 0.2, 10)
# all the data points
X = concatenate((X_1, X_2), axis=1)
group = array([1] * 10 + [2] * 10)
g1 = (group == 1)
g2 = (group == 2)

scatter(array(X[0, g1]), array(X[1, g1]), color='b')
scatter(array(X[0, g2]), array(X[1, g2]), color='r')
axis([-2, 2, -2, 2])
```
